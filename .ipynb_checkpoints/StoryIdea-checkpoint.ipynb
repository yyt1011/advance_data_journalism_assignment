```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('campus_crime.csv')
```

```python
data[:1]
```

```python
data["total_count"] = data["murder"] + data["manslaughter"] + data["forcible"]\
  + data["nonforce"] + data["robbery"] + data["agg_aslt"] + data["burglary"]\
  + data['vehicle'] + data["arson"]
```

```python
g = sns.FacetGrid(data, size=5)
g.map(plt.scatter, "total_enr", "total_count", alpha=.8)
```

```python
g = sns.FacetGrid(data,
                  hue='murder', 
                  size=5)
g.map(plt.scatter, "total_enr", "total_count", 
      alpha=.8, 
      s=60)
g.add_legend()
```

```python
small_totalenr = data[data.total_enr<= 200000]
g=sns.FacetGrid(small_totalenr, 
                hue='murder',
                size=5)
g.map(plt.scatter, "total_enr", "total_count", 
      alpha=.8, 
      s=60)
g.add_legend()
```

```python
data.sort_values(by='total_count',ascending=False)[:1]
```

```python
g = sns.FacetGrid(data[data.total_enr>280000], size=5)
g.map(plt.scatter, 'total_enr', 'total_count', alpha=.8, s=60)
```

```python
data.sort_values(by='total_enr',ascending=False)[:20]
```

# Story Idea 1: 
### Arizona State University, though one of the universities with a large number of students, has the highest number of crime. 
### Among other universities with larger student populations, a bigger number of crime at Arizona State University can also be obversed. 
### Among the top 20 universities in terms of total crimes, arson, forcible, burglary,and aggressive assault are reported more at Arizona State University than at the rest of the universities. 
### Considering that violence is more likely to be used in these crimes, students and teachers should pay more attention to their personal security. Thus, what measures did the university take? Are the students well-educated of how to protect themselves?

```python
g = sns.PairGrid(data=small_totalenr,vars=['robbery','agg_aslt'],size=5)
g.map_offdiag(plt.scatter);
```

```python
g = sns.lmplot(data=small_totalenr, x='robbery', y='agg_aslt',size=5)
sns.regplot(x="robbery", y="agg_aslt", data=small_totalenr, scatter=False, ax=g.axes[0, 0])
```

# Story Idea 2:
## There is a positive relation between the number of robberies and the number of aggressive assaults. Given that both crimes have the potential to cause injury, it's also worth knowing how many students have been hurt in these cases and whethere are there overlaps bewteen assaulters and robbers.

```python
g = sns.PairGrid(data=small_totalenr,vars=['total_count','burglary'],size=5)
g.map_offdiag(plt.scatter);
```

```python
g = sns.lmplot(data=small_totalenr, x='total_count', y='burglary',size=5)
sns.regplot(x="total_count", y="burglary", data=small_totalenr, scatter=False, ax=g.axes[0, 0])
```
